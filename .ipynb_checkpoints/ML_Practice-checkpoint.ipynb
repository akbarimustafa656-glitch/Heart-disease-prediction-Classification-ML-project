```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
import joblib
import pickle
from sklearn.ensemble import HistGradientBoostingRegressor
from sklearn.linear_model import Ridge
```

```python
np.random.seed(420)
hd = pd.read_csv("sk-data/heart-disease.csv")
x = hd.drop("target", axis=1)
y = hd["target"]
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
heart_disease = RandomForestClassifier()
heart_disease.fit(x_train, y_train)
pickle.dump(heart_disease, open("Heart_disease_predection.pkl", "wb"))
```

```python
# used RandomForrestRgressior
np.random.seed(12)
cardekho = pd.read_csv("sk-data/cardekho.csv")
cardekho["name"].fillna("missing", inplace=True)
cardekho["year"].fillna(2014, inplace=True)
cardekho["km_driven"].fillna(cardekho["km_driven"].mean(), inplace=True)
cardekho["fuel"].fillna("missing", inplace=True)
cardekho["seller_type"].fillna("missing", inplace=True)
cardekho["transmission"].fillna("missing", inplace=True)
cardekho["owner"].fillna("missing", inplace=True)
cardekho["mileage(km/ltr/kg)"].fillna(cardekho["mileage(km/ltr/kg)"].mean(), inplace=True)
cardekho["engine"].fillna(cardekho["engine"].mean(), inplace=True)
cardekho["max_power"] = pd.to_numeric(cardekho["max_power"], errors="coerce")
cardekho["max_power"].fillna(cardekho["max_power"].mean(), inplace=True)
cardekho["seats"].fillna(5, inplace=True)
cardekho.dropna(inplace=True)
x = cardekho.drop("selling_price", axis=1)
y = cardekho["selling_price"]
```

```python
cgfs = ["name", "fuel", "seller_type", "transmission", "owner"]
one_hot = OneHotEncoder()
transformer = ColumnTransformer([("one_hot",
                                 one_hot,
                                 cgfs)],
                               remainder="passthrough")
transformed_x = transformer.fit_transform(x)
x_train, x_test, y_train, y_test = train_test_split(transformed_x, y, test_size=0.2)
```

```python
np.random.seed(1399)
car = RandomForestRegressor()
# x_train = x_train.toarray()
# x_test = x_test.toarray()
car.fit(x_train, y_train)
```

```python
car.score(x_test, y_test)
```

```python
pickle.dump(car, open("car_price_predection.pkl", "wb"))
```

```python
# Data science
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
# Classification Models training
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.ensemble import HistGradientBoostingClassifier
from sklearn.ensemble import ExtraTreesClassifier
from sklearn.linear_model import LogisticRegression
# Saving trained models
import pickle
import joblib
```

# Using RandomForestClassifier
for heart-disease.csv

```python
np.random.seed(101)
df = pd.read_csv("SK-data/heart-disease.csv")
x = df.drop("target", axis=1)
y = df["target"]
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
models = {"Heart_disease_Randoom_Forrest": RandomForestClassifier(),
         "Heart_disease_Logistic": LogisticRegression(),
         "Heart_disease_KNN": KNeighborsClassifier(),
         "Heart_disease_Hist": HistGradientBoostingClassifier(),
         "Heart_disease_ExtraTree": ExtraTreesClassifier(n_estimators=100),
         "Heart_disease_MLP": MLPClassifier()}
models_score = {}
for name, model in models.items():
    model.fit(x_train, y_train)
    models_score[name] = model.score(x_test, y_test)
models_score
```

## Saving Our Classification Trained Models

```python
for name, model in models.items():
    pickle.dump(model, open(f"{name}", "wb"))
```

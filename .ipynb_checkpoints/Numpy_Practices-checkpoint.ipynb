```python
metrics = np.array([[2,7,1],
                   [9,4,16],
                   [11,14,18],
                   [13,13,16],
                   [15,18,9]])

prices = np.array([10,8,12])
```

```python
metrics * prices
```

```python
result = prices.dot(metrics.T)
result
```

```python
daily = pd.DataFrame(metrics,
                    index=["Mon", "Tus", "Wed", "Thurs", "Fri"],
                    columns=["Burger", "Bolani", "Filafil"])
Main_prices = pd.DataFrame(prices.reshape(1,3),
                          index=["Prices"],
                          columns=["Burger", "Bolani", "Filafil"])
daily["Total $"] = result.T
daily
```

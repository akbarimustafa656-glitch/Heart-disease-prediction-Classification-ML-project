```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
np.random.seed(10)
hd = pd.read_csv("sk-data/heart-disease.csv")
x = hd.drop("target", axis=1)
y = hd["target"]
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.2)
clf = RandomForestClassifier()
clf.fit(x_train, y_train)
lis = []
elements = ['age', 'gender', 'cp', 'trestbps', 'chol', 'fbs', 'restecg', 'thalach',
       'exang', 'oldpeak', 'slope', 'ca', 'thal']
for i in elements:
    enteres = float(input(f"Plese enter your information about {i}: "))
    lis.append(enteres)
np_array = np.array(lis)
prediction = clf.predict(np_array.reshape(1,-1))
if prediction == 0:
    print("Congrats! you don't have a heart problem.")
else:
    print("Be careful! you might have a heart problem.")
```

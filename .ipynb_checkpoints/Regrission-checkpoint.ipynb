```python
# Data Science
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
# Regression models training
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import ExtraTreesRegressor
from sklearn.ensemble import HistGradientBoostingRegressor
from sklearn.linear_model import Ridge
from sklearn.neighbors import KNeighborsRegressor
from sklearn.neural_network import MLPRegressor
# Saving models
import pickle
import joblib
```

## Car price predection project with for car-sales-extended-missing-data.csv Dataset

## Using RandomForrestRegressior
for car-sales-extended-missing-data

```python
np.random.seed(890)
cars = pd.read_csv("sk-data/car-sales-extended-missing-data.csv")
cars["Make"].fillna("missing", inplace=True)
cars["Colour"].fillna("missing", inplace=True)
cars["Odometer (KM)"].fillna(cars["Odometer (KM)"].mean(), inplace=True)
cars["Doors"].fillna(4, inplace=True)
cars.dropna(inplace=True)
x = cars.drop("Price", axis=1)
y = cars["Price"]
cgfs = ["Make", "Colour", "Doors"]
one_hot = OneHotEncoder()
transformer = ColumnTransformer([("one_hot",
                                 one_hot,
                                 cgfs)],
                               remainder = "passthrough")
transformed_x = transformer.fit_transform(x)
x_train, x_test, y_train, y_test = train_test_split(transformed_x, y, test_size=0.2)
models = {"First_car_price_RandomForrest": RandomForestRegressor(n_jobs=-1),
         "First_car_price_Hist": HistGradientBoostingRegressor(),
         "First_car_price_KNN": KNeighborsRegressor(n_jobs=-1),
         "First_car_price_ExtraTree": ExtraTreesRegressor(n_estimators=100),
         "Frist_car_price_Ridge": Ridge(),
         "First_car_price_MLP": MLPRegressor()}
x_train = x_train.toarray()
x_test = x_test.toarray()
for name, model in models.items():
    model.fit(x_train, y_train)
```

## Car price predection project with cardekho.csv Dataset

```python
np.random.seed(123456789)
card = pd.read_csv("sk-data/cardekho.csv")
card["mileage(km/ltr/kg)"].fillna(card["mileage(km/ltr/kg)"].mean(), inplace=True)
card["engine"].fillna(card["engine"].mean(), inplace=True)
card["seats"].fillna(card["seats"].mean(), inplace=True)
card["max_power"] = pd.to_numeric(card["max_power"], errors="coerce")
card["max_power"].fillna(card["max_power"].mean(), inplace=True)
x = card.drop("selling_price", axis=1)
y = card["selling_price"]
cgfs = ["name", "fuel", "seller_type", "transmission", "owner"]
one_hot = OneHotEncoder()
transformer = ColumnTransformer([("one_hot",
                                 one_hot,
                                 cgfs)],
                               remainder = "passthrough")
transformed_x = transformer.fit_transform(x)
x_train, x_test, y_train, y_test = train_test_split(transformed_x, y, test_size=0.2)
x_train = x_train.toarray()
x_test = x_test.toarray()
models1 = {"second_car_price_RandomForrest": RandomForestRegressor(n_jobs=-1),
         "Second_car_price_KNN": KNeighborsRegressor(n_jobs=-1),
         "Second_car_price_MLP": MLPRegressor(),
         "Second_car_price_Hist": HistGradientBoostingRegressor(),
         "Second_car_price_Ridge": Ridge(),
         "Second_car_price_ExtraTree": ExtraTreesRegressor(n_estimators=100)}
for name, model in models1.items():
    model.fit(x_train, y_train)
```

## Saving our trained models

### Save models (cardekho.csv) Dataset

```python
for name, model in models1.items():
    pickle.dump(model, open(f"{name}.pkl", "wb"))
```

### Save models (car-sales-extended-missing-data)

```python
for name, model in models.items():
    pickle.dump(model, open(f"{name}.pkl", "wb"))
```
